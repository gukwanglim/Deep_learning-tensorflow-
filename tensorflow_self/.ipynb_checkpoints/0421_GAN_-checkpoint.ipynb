```python
import tensorflow as tf
from tensorflow import keras

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os

from sklearn.model_selection import train_test_split
```

```python
## 생성자 만들기

generator = keras.Sequential()
generator.add(keras.layers.Dense(7*7*128, input_dim = 100,                         # input_dim은 임의의 랜덤 벡터를 넣어줌
                                activation = keras.layers.LeakyReLU(alpha = 0.2)))
generator.add(keras.layers.BatchNormalization())         # BatchNormalization : 평균이 0, 분산은 1이 되도록 만들어줌.
generator.add(keras.layers.Reshape((7, 7, 128)))
generator.add(keras.layers.UpSampling2D())
generator.add(keras.layers.Conv2D(64, kernel_size = 5, padding = 'same'))

generator.add(keras.layers.BatchNormalization())
generator.add(keras.layers.Activation(keras.layers.LeakyReLU(0.2)))

generator.add(keras.layers.UpSampling2D())
generator.add(keras.layers.Conv2D(1, kernel_size = 5, padding = 'same', activation = 'tanh')) 
                                                      # padding 이 없으면 (24, 24)로 크기가 줄어서 나온다.
generator.summary()
```

- <span style = 'color:blue'>input_dim = 100</span> : 100차원 크기의 랜덤 벡터를 준비해 집어 넣어라
- <span style = 'color:blue'>UpSampling2D</span> : 이미지의 가로, 세로 크기를 2배씩 늘려줌.
- <span style = 'color:blue'>Reshape</span> : 컨볼루션 레이어가 받아들일 수 있는 형태로 바꿔주는 코드. (Conv2D() 함수의 input_shape 부분에 들어갈 형태로 정해줌)
- <span style = 'color:blue'>LeakyReLU</span> : GAN에서는 relu 대신에 LeakyReLU를 사용.(마지막 layer에서는 tanh)
- <span style = 'color:blue'>BatchNormalization</span> : 평균이 0, 분산은 1인 모습으로 만들어줌.(배치 정규화)(Z transform)

$$ z = \frac{x_i - \bar{x}}{\sigma} $$

마지막 컨볼루션을 거치고 <span style = 'color:blue'>(28, 28, 1)의 형태로 만들어 판별자로 넘길 준비.</span>

<span style = 'color:red'>생성자는 compile을 하지 않는다.</span>

```python
## 판별자 만들기

discriminator = keras.Sequential()
discriminator.add(keras.layers.Conv2D(64, kernel_size = 5, strides = 2, input_shape=(28, 28, 1), padding = 'same'))
discriminator.add(keras.layers.Activation(keras.layers.LeakyReLU(0.2)))
discriminator.add(keras.layers.Conv2D(128, kernel_size = 5, strides = 2, padding = 'same'))
discriminator.add(keras.layers.Activation(keras.layers.LeakyReLU(0.2)))
discriminator.add(keras.layers.Flatten())
discriminator.add(keras.layers.Dense(1, activation = 'sigmoid'))

discriminator.compile(loss = 'binary_crossentropy',
                     optimizer = 'adam')

discriminator.trainable = False
discriminator.summary()
```

- <span style = 'color:blue'>LeakyReLU</span> : 생성자와 마찬가지로 판단자도 LeakyReLU 사용.
- <span style = 'color:blue'>Flatten</span> : 판별은 Dense로 하기 때문에 평탄화 작업이 필요하다.
- <span style = 'color:blue'>discriminator.trainable = False</span> : <span style = 'color:red'>판별자 자신이 학습되지 않도록</span> 학습 기능을 off.

```python
## 생성자와 판별자의 모델을 연결시키는 gan 모델 만들기

ginput = keras.layers.Input(shape = (100,))                 # 생성자에서 input_dim = 100 이었으므로 Input(shape = (100, ))

dis_output = discriminator(generator(ginput))
gan = keras.models.Model(ginput, dis_output)

gan.compile(loss = 'binary_crossentropy',
           optimizer = 'adam')

gan.summary()
```

- <span style = 'color:blue'>Input(shape = (100,))</span> : 랜덤한 100개의 벡터를 keras.Input() 함수에 집어 넣어 <span style = 'color:red'>생성자에 입력</span>할 ginput을 만드는 과정.
- <span style = 'color:blue'>dis_output = discriminator(generator(ginput))</span> : 생성자 모델에 ginput을 입력.(결과 출력은 28*28 크기의 이미지가 그대로 <span style = 'color:red'>판별자 모델의 입력값</span>에 들어감)
- <span style = 'color:blue'>Model</span> : ginput에서 구한 값과 dis_output의 값을 넣어 gan이라는 새로운 모델 생성.

```python
def gan_train(epoch, batch_size, saving_interval):
    
    (x_train, _), (_,_) = keras.datasets.mnist.load_data()           # 테스트 과정은 필요없고 이미지만 필요
    x_train = x_train.reshape(-1, 28, 28, 1).astype(np.float32)
    x_train = (x_train - 127.5)/127.5                               # -1 ~ 1 사이의 값을 만들기 위해 중간값을 뺀 후, 나누기
    
    true = np.ones((batch_size, 1))      # 모두 참             # np.ones(())에서 () 안의 ()는 shape를 의미. 즉, 입력된 batch_size
    fake = np.zeros((batch_size, 1))     # 모두 거짓           # 만큼의 행과 1개의 열을 만들어 0, 1을 채우겠다는 의미.
    
    d_losses = []                         # d는 판별자
    g_losses = []                         # g는 생성자
    
    for i in range(epoch):
        
        # real training loss
        idx = np.random.randint(0, x_train.shape[0], batch_size)  # 실제 이미지에서 저장
        imgs = x_train[idx]
        
        d_loss_real = discriminator.train_on_batch(imgs, true)         # 판별자 모델에 train_on_batch() 함수를 사용하여 판별
        
        # generator loss
        noise = np.random.normal(0, 1, (batch_size, 100))            # 생성자에 집어 넣을 가짜 이미지
        gen_imgs = generator.predict(noise)                       # 가상의 이미지 값이 생성자에 들어가고 결괏값을 저장
        d_loss_fake = discriminator.train_on_batch(gen_imgs, fake)  # gen_imgs는 모두 가짜이다.
        
        d_loss = np.add(d_loss_real, d_loss_fake)/2.               # d_loss_real과 d_loss_fake의 평균 = d_loss
        g_loss = gan.train_on_batch(noise, true)                  # gan 모델(위에서 만듦)을 이용해 생성자의 오차를 구함.
        
        d_losses.append(d_loss)
        g_losses.append(g_loss)
        
        print('epoch = ', i, 'd_loss = ', d_loss, 'g_loss = ', g_loss)
        
        if i % saving_interval == 0:                              # 여기서 savig_interval는 input 값
            noise = np.random.normal(0, 1, (25, 100))
            gen_images = generator.predict(noise)
            
            gen_images = gen_images*0.5 + 0.5             # generator가 tanh(-1 ~ 1)이므로 0.5를 곱하고 다시 0.5를 더해 양의 값으로 변환
            
            fig, axe = plt.subplots(5, 5)
            count = 0
            for j in range(5):
                for k in range(5):
                    axe[j, k].imshow(gen_imges[count, :, :, 0], cmap = 'gray')
                    axe[j, k].axis('off')
                    
                    count += 1
            fig.savefig('gan_images/gan_mnist_%d.png' % i)
            
    return d_losses, g_losses
```

실제 이미지를 가지고 판별

- <span style = 'color:blue'>true와 fake에서 batch_size</span> : 한 번에 몇 개의 실제 이미지와 몇 개의 가상 이미지를 판별자에 넣을지 결정.
    - <span style = 'color:blue'>np.zeros/ones()</span> : 여기서 ones(())일 경우, () 안의 ()는 shape를 의미.즉, 위 코드에서는 입력된 batch_size만큼의 행과 1개의 열을 만들어 0, 1을 채우겠다는 의미.
- <span style = 'color:blue'>np.random.randint()</span> : 실제 이미지를 랜덤하게 선택해서 볼러옴.(a, b, c)의 경우 a부터 b까지의 숫자 중 하나를 랜덤하게 선택해 가져오는 과정을 <span style = 'color:red'>c번 반복</span>하라는 뜻.
- <span style = 'color:blue'>discriminator.train_on_batch(imgs, true)</span> : 판별자 모델(discriminator)에서 train_on_batch() 함수를 사용하여 <span style = 'color:red'>판별 시작</span>. train_on_batch(x, y)는 입력 값(x)과 레이블(y)를 받아서 <span style = 'color:red'>딱 한 번 학습을 실시</span>해 모델을 업데이트. 위 코드에서는 바로 윗줄에서 만든 imgs를 x에 넣고, true 배열을 y에 놓아 준비.

가짜 이미지를 가지고 판별

- <span style = 'color:blue'>noise = np.random.normal(0, 1, (batch_size, 100))</span> : 생성자에 넣을 가상의 이미지는 정수가 아니기 때문에 normal() 함수를 사용. random() 함수처럼 (a, b, c)로 a부터 b까지 c번 반복이라는 의미. 위 코드에서 c 자리에 있는 (batch_size, 100)은 batch_size 만큼 100열을 뽑으라는 의미.
- <span style = 'color:blue'>d_loss_fake = discriminator.train_on_batch(gen_imgs, fake)</span> :  판별자 모델(discriminator)에서 train_on_batch() 함수를 사용하여 <span style = 'color:red'>판별 시작</span>.
    - for 구문 안에서 돌아가는 코드이므로 d_loss_real과 d_loss_fake가 판별자 안에서 번갈아가며 진위를 판단.
    
- <span style = 'color:blue'>d_loss = np.add(d_loss_real, d_loss_fake)/2.</span> : d_loss_real과 d_loss_fake의 평균을 구해서 저장.
- <span style = 'color:blue'>g_loss = gan.train_on_batch(noise, true)</span> : 판별자와 생성자를 연결하여 만든 gan 모델을 이용해 생성자의 오차 g_loss를 구함. 여기서는 train_on_batch(noise, true)으로 생성자의 레이블(noise)를 무조건 참이라며 판별자에게 넘김.


```python
d_loss, g_loss = gan_train(4001, 32, 200)
```
