```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-2, 2, 0.01)
print(x)
```

```python
y = np.log2(x)
y1 = -(np.log2(x))
y2 = np.log2(-x)
y3 = -(np.log2(-x))
y4 = -(np.log2(1-x))
```

```python
plt.figure(figsize = (6, 6))
plt.plot(x, y, label = 'log(x)')
plt.plot(x, y1, label = '-log(x)')
plt.plot(x, y2, label = 'log(-x)')
plt.plot(x, y3, label = '-(log(-x))')
plt.plot(x, y4, label = '-(log(1-x))')

plt.axhline(0, c ='k')                                            # grid 축 색칠하기 (y축)
plt.axvline(0, c ='k')                                            # grid 축 색칠하기 (x축)

plt.grid()
plt.ylim(-5, 5)
plt.legend()

plt.show()
```

```python
plt.figure(figsize = (6, 6))
plt.plot(x, y1, label = '-log(x)')
plt.plot(x, y4, label = '-(log(1-x))')

plt.ylim(0, 5)
plt.xlim(0, 2)
plt.legend()

plt.show()
```
