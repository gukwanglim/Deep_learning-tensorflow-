# 14장 베스트 모델 만들기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
```

```python
df_pre = pd.read_csv('../교재 코드/deeplearning-for-everyone-2nd-master/dataset/wine.csv',
                header = None)
```

```python
df_pre.head()
```

```python
df_pre.info()
```

```python
df = df_pre.sample(frac = 0.5)                         # sample() 함수는 원본 데이터에서 정해진 비율만큼 랜덤으로 뽑아옴.
                                                    # frac = 1 일 경우 데이터의 100%를 가져오라는 뜻
                                                    # frac = 0.5 일 경우 데이터의 50%를 가져오라는 뜻.
```

```python
df.shape
```

```python
dataset = df.values
dataset
```

```python
x = dataset[:, :-1]
y = dataset[:, -1].astype(np.int16)

print(x.dtype)
print(y.dtype); print()

y[:10]
```

```python
model = keras.Sequential()                                                       # 모델 쌓기
model.add(keras.layers.Dense(30, input_dim = 12, activation = 'relu'))
model.add(keras.layers.Dense(12, activation = 'relu'))
model.add(keras.layers.Dense(8, activation = 'relu'))
model.add(keras.layers.Dense(1, activation = 'sigmoid'))

model.summary()
```

```python
model.compile(loss = 'binary_crossentropy',
             optimizer = 'adam',
             metrics = 'accuracy')
```

```python
history = model.fit(x, y, epochs = 200, batch_size = 100)
```

## 에포크마다 모델의 정확도를 함께 기록하며 저장하기

```python
import os

model_dir = './model/'                               # 모델을 저장하는 폴더 선택

if not os.path.exists(model_dir):                     # 만약 모델을 저장하는 폴더자 존재하지 않을 경우
    os.mkdir(model_dir)                               # () 안의 이름을 가진 폴더를 만들어라.
```

```python
modelpath = './model/{epoch:02d}-{val_loss:.4f}.hdf5'                   # 01, 02, 03 으로 저장, 소수점 넷째 자리까지(파일 이름 설정)
checkpoint = keras.callbacks.ModelCheckpoint(filepath = modelpath,  # ModelCheckpoint() : 모델을 저장하기 위한 함수
                                            monitor = 'val_loss',   # 모니터할 값을 지정.
                                            verbose = 1,            # 해당 함수의 진행사항 출력, 만약 0으로 설정할 경우 나오지 않음.
                                            save_best_only = True) # 좋은 것만 세이브 

history = model.fit(x, y, validation_split=0.2, epochs = 200, batch_size = 200,
         verbose = 1, callbacks = [checkpoint])                    # callbacks는 리스트의 형태로 지정(지정받은 곳에 저장)
```

- ModelCheckpoint() : 모델을 저장하기 위한 함수. keras.callbacks 아래에 위치.

    - filepath : 모델이 저장될 장소.
    - save_best_only : 제일 좋은 데이터를 수집

- 테스트 오차는 keras 내부에서 val_loss로 기록.(정확도 = acc, 테스트셋 정확도 = val_acc, 학습셋 오차 = loss로 기록)



- validation_split = 0.2 : 전체의 데이터 중 20%는 테스트 세트로 만들어라.

```python
df = df_pre

x = df.values[:, :-1]
y = df.values[:, -1].astype(np.int16)
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(30, input_dim = 12, activation = 'relu'))
model.add(keras.layers.Dense(12, activation = 'relu'))
model.add(keras.layers.Dense(8, activation = 'relu'))
model.add(keras.layers.Dense(1, activation = 'sigmoid'))

model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', 
             optimizer = 'adam',
             metrics = 'accuracy')
```

```python
modelpath = './model/{epoch:02d}-{val_loss:.4f}.hdf5'

checkpoint = keras.callbacks.ModelCheckpoint(filepath = modelpath,
                                            monitor = 'val_loss',
                                            verbose = 1,
                                            save_best_only = True)

early_stopping = keras.callbacks.EarlyStopping(monitor = 'val_loss',            # 데이터가 더 이상의 효율이 나오지 않으면 멈춰라
                                              patience = 10)                   # patience : 다만, 이정도까지는 돌리고 멈춰라

history = model.fit(x, y, validation_split=0.3, epochs = 1000,
                   batch_size = 500, verbose = 1,
                   callbacks = [checkpoint, early_stopping])
```

- EarlyStopping : 데이터가 더 이상의 효율이 나오지 않으면 멈추게 하는 함수. keras.callbacks 아래에 있다.
    - patience : 다만, patience에서 정해진 횟수만큼은 돌고 멈춰라.

```python
val_loss = history.history['val_loss']
val_acc = history.history['val_accuracy']

plt.plot(val_loss, c ='r')
plt.plot(val_acc, c = 'b')
plt.show()
```

```python
dataset = df_1.values

X = dataset[:, :-1]
Y = dataset[:, -1]
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(30, input_dim = 12, activation = 'relu'))
model.add(keras.layers.Dense(12, activation = 'relu'))
model.add(keras.layers.Dense(8, activation = 'relu'))
model.add(keras.layers.Dense(1, activation = 'sigmoid'))

model.summary()
```

```python
MODEL_DIR = './model_test/'

if not os.path.exists(MODEL_DIR):
    os.mkdir(MODEL_DIR)
    
modelpath = './model_test/{epoch:02d}-{val_loss:.4f}.hdf5'

checkpointer = keras.callbacks.ModelCheckpoint(filepath = modelpath,
                                              monitor = 'val_loss',
                                              verbose = 1, 
                                              save_best_only = True)

history_1 = model.fit(X, Y, validation_split=0.33, epochs = 3500, batch_size = 500,
                     verbose = 1, callbacks = [checkpointer])      # callbacks = [checkpointer] 를 사용하지 않으면 저장되지 않는다.

y_vloss = history_1.history['val_loss']
y_acc = history_1.history['accuracy']

x_len = np.arange(len(y_acc))
plt.plot(x_len, y_vloss, 'o', c ='r', markersize = 3)
plt.plot(x_len, y_acc, 'o', c ='b', markersize = 3)

plt.show()
```
